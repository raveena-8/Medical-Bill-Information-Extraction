```python
import fitz  # PyMuPDF

# Open the PDF file
pdf_path = '/content/Sample_For_Assignment.pdf'
pdf_document = fitz.open(pdf_path)

# Iterate through each page
for page_num in range(pdf_document.page_count):
    page = pdf_document.load_page(page_num)

    # Create a new PDF document for each page
    new_pdf = fitz.open()
    new_pdf.insert_pdf(pdf_document, from_page=page_num, to_page=page_num)

    # Save each page as a separate PDF file
    output_path = f'bill_{page_num + 1}.pdf'
    new_pdf.save(output_path)
    new_pdf.close()

pdf_document.close()
```

```python
import pdfplumber

# Example function to extract text from a bill
def extract_information(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Use the function for each bill PDF
for i in range(1, 5):  # Change range depending on the number of bills
    bill_path = f'bill_{i}.pdf'
    bill_text = extract_information(bill_path)
    # Process the extracted text as needed
    print(f"Information from bill {i}:\n{bill_text}")
```

```python
from pdf2image import convert_from_path
from pytesseract import image_to_string
```

```python
import json
import re
```

```python
def convert_pdf_to_img(pdf_file):
    return convert_from_path(pdf_file)

def convert_image_to_text(file):
    text= image_to_string(file)
    return text

def get_text_from_any_pdf(pdf_file):
    images = convert_pdf_to_img(pdf_file)
    final_text=""
    for pg, img in enumerate(images):

        final_text += convert_image_to_text(img)

    return final_text
```

```python
def extract_header_info(text):
    header_info = {}

    # Regular expressions for pattern matching
    patterns = {
        "hospital_clinic": r"(Singapore\s*General\s*Hospital|Tan\s*To\S*k\s*Seng\s*Hospital)",
        "gst_reg_no": r"GST\s*REG\s*NO\s*:\s*([\w-]+)",
        "visit_date": r"Visit\s*Date\s*:\s*(\d{2}.\d{2}.\d{4})",
        "invoice_date": r"Tax\s*Invoice\s*Date\s*—\s*:\s*(\d{2}.\d{2}.\d{4})",
        "bill_receipt_date": r"Bill\s*Ref\s*Number\s*[:=]\s*(\d{4}-\d{4}-\d{4}-\w{4}-\d{2})",
        "admission_date": r"Admission\s*Date\s*:\s*(\d{2}.\d{2}.\d{4})",
        "discharge_date": r"Discharge\s*Date\s*:\s*(\d{2}.\d{2}.\d{4})",
        "invoice_number": r"Tax\s*Invoice\s*Number\s*:\s*(\d{10}[A-Za-z0-9]*)",
        "total_amount_payable": r"NET\s*AMOUNT\s*PAYABLE\s*I\s*C\s*(\d+\.\d{2})",
        "invoice_page": r"Page\s*(\d+)\s*of\s*(\d+)",
        "doctor_name": r"DR\s*\.\s*(\w+\s*\w+)",
        "location": r"Location\s*:\s*(\w+)",
        "bill_type": r"(ORIGINAL|DUPLICATE|INTERIM)"

    }

    # Extract information based on patterns
    for key, pattern in patterns.items():
        match = re.search(pattern, text)
        if match:
            header_info[key] = match.group(1)

    return header_info

# Function to extract line-item information (table data)
def extract_line_items(text):
    line_items = []

    # Regular expression for line-item information
    line_item_pattern = r"(\d+)\s+(.+?)\s+(\d+)\s+([\d,.]+)"

    matches = re.finditer(line_item_pattern, text)
    for match in matches:
        line_items.append({
            "Code": match.group(1),
            "Description": match.group(2),
            "Quantity": match.group(3),
            "Amount": match.group(4)
            # Add more fields if available in the text
        })

    return line_items


```

```python
def process_multiple_bills(bill_paths):
    for i, path_to_pdf in enumerate(bill_paths, start=1):
        invoice_text = get_text_from_any_pdf(path_to_pdf)
        header_info = extract_header_info(invoice_text)
        line_items = extract_line_items(invoice_text)

        # Combine extracted information into a JSON format
        structured_data = {
            f"HeaderInformation_Bill_{i}": header_info,
            f"LineItemInformation_Bill_{i}": line_items
        }

        # Convert structured data to JSON and print
        json_data = json.dumps(structured_data, indent=4)
        print(f"Information for Bill {i}:\n{json_data}")

# List of paths to each bill PDF
bill_files = ['/content/bill_1.pdf', '/content/bill_2.pdf', '/content/bill_3.pdf', '/content/bill_4.pdf']

# Process all the bills
process_multiple_bills(bill_files)
```

**Libraries Used**
1. pdf2image: Used for converting PDF pages to images for OCR.
2. pytesseract: Employed for OCR (optical character recognition) to extract text from images.
3. Regular Expressions (regex): Utilized for pattern matching and text extraction.
4. PyMuPDF: For PDF parsing and text extraction.
5. pdfplumber: For extracting table data and structured information from PDFs.
6. poppler-utils: External tool used by pdfplumber for efficient PDF parsing.

**Usage**

Input: Multi-page PDF medical bills.

Output: JSON format containing extracted information (header, table data, etc.).

**Code Overview**

1. The code uses PyMuPDF to split the multi-page PDF into individual bills.
2. It employs pdfplumber and pytesseract for text extraction and table data extraction from each bill.
3. The extracted information is formatted into JSON for each bill separately.

**Instructions**

2. Update the bill_files list with the correct paths to your bill files.
